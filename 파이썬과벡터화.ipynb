### 행렬의 네 열 안의 수의 합을 구하고 행렬 전체를 나눠서 네가지 음식 안의 탄단지가 주는 칼로리의 백분율을 구한다
- for 문을 쓰지 않고 한다


```python
# 행렬 생성
import numpy as np
A= np.array([[56.0,0.0,4.4,68.0],
            [1.2,104.0,52.0,8.0],
            [1.8,135.0,99.0,0.9]])
print(A)
```

```python
# 각 열마다 총합 구하기
cal=A.sum(axis=0) # 세로로 더해라 가로축은 axis 0 이다
cal2 = A.sum(axis=1) # 가로로 더해라
print(cal)
print(cal2)
```

```python
# 백분율 행렬 만들기
print(cal.reshape(1,4)) #원래 1차원 행렬을 1by4 2차원 행렬로 만들어준다 근데 여기서 굳이 할 필요는 없음
# reshape 은 상수 시간이 소요된다.
percentage = 100* A/cal#.reshape(1,4)
print(percentage)
```

```python
B=np.array([1,2,3,4])
print(B+100) # 100을 자동으로 4*1 벡터로 만들어주어 각각의 원소에 대해 연산 가능하게한다.
```

# 파이썬과 넘파이 벡터
## 파이썬의 유연성은 장단점이 될 수 있다
- 단점: 잘 모르면 가끔 이상하고 찾기 어려운 에러가 발생한다. 에러가 안 뜨고 예상 못 한 값이 나올 수 있움..

```python
import numpy as np

a= np.random.randn(5) # 가우시안 분포를 따르는 변숫값 5개를 배열 a에 저장한다.
print(a)
```

```python
print(a.shape) # rank 1 array , 행벡터 열벡터도 아님 -> 직관적이지 않은 결과를 도출한다
```

```python
print(a.T)
print(a.T.shape)
```

```python
print(np.dot(a,a.T)) # 행렬이 나와야하는데 상숫값이 나온다
```

## rank 1 array 는 아예 사용하지 마라
- 결과가 직관적이지 않음

```python
a=np.random.randn(5,1)
print(a) # 괄호가 두개임
print(a.shape)
```

```python
print(a.T)
```

```python
print(np.dot(a,a.T)) # 벡터의 외적 => 행렬
print(np.dot(a.T,a))
```

```python
assert(a.shape==(5,1)) # 검증 코드
```
